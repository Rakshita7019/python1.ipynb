```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate a dataset with outliers
np.random.seed(0)  # For reproducibility
data = np.random.normal(0, 1, 100)  # Generate 100 normal data points
data = np.append(data, [10, -10])  # Add outliers

# Calculate the mean, median, and mode
mean = np.mean(data)
median = np.median(data)

# Correct mode calculation
mode_result = stats.mode(np.round(data))  # Get ModeResult object

# Handle mode result depending on whether it's a scalar or array
if isinstance(mode_result.mode, np.ndarray):  # When there are multiple modes
    if mode_result.count[0] > 0:  # If the mode is valid (appears in the data)
        mode = mode_result.mode[0]  # The mode is the first element in the 'mode' array
    else:
        mode = None  # No valid mode
else:  # When there is a single mode
    mode = mode_result.mode  # Use the mode directly

# Print the results
print(f"Mean: {mean:.4f}")
print(f"Median: {median:.4f}")
print(f"Mode: {mode:.4f}" if mode is not None else "Mode: No mode found")

# Visualize the data with outliers
plt.hist(data, bins=30, density=True, alpha=0.6)
plt.axvline(mean, color='r', label='Mean')
plt.axvline(median, color='g', label='Median')
plt.axvline(mode, color='b', label='Mode' if mode is not None else 'Mode: No mode')
plt.legend()
plt.title("Distribution with Outliers")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# Remove the outliers (values greater than 5 or less than -5)
data_no_outliers = data[np.abs(data) < 5]

# Recalculate mean, median, and mode without outliers
mean_no_outliers = np.mean(data_no_outliers)
median_no_outliers = np.median(data_no_outliers)

# Correct mode calculation again for the data without outliers
mode_no_outliers_result = stats.mode(np.round(data_no_outliers))  # Get ModeResult object

# Handle mode result depending on whether it's a scalar or array for data without outliers
if isinstance(mode_no_outliers_result.mode, np.ndarray):  # When there are multiple modes
    if mode_no_outliers_result.count[0] > 0:  # If the mode is valid (appears in the data)
        mode_no_outliers = mode_no_outliers_result.mode[0]  # The mode is the first element in the 'mode' array
    else:
        mode_no_outliers = None  # No valid mode
else:  # When there is a single mode
    mode_no_outliers = mode_no_outliers_result.mode  # Use the mode directly

# Print the recalculated results without outliers
print(f"Mean (without outliers): {mean_no_outliers:.4f}")
print(f"Median (without outliers): {median_no_outliers:.4f}")
print(f"Mode (without outliers): {mode_no_outliers:.4f}" if mode_no_outliers is not None else "Mode (without outliers): No mode found")

# Visualize the data without outliers
plt.hist(data_no_outliers, bins=30, density=True, alpha=0.6)
plt.axvline(mean_no_outliers, color='r', label='Mean')
plt.axvline(median_no_outliers, color='g', label='Median')
plt.axvline(mode_no_outliers, color='b', label='Mode' if mode_no_outliers is not None else 'Mode: No mode')
plt.legend()
plt.title("Distribution without Outliers")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Import the necessary modules
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Create two datasets: scores before and after the program
np.random.seed(0)  # For reproducibility
scores_before = np.random.normal(80, 10, 100)  # 100 scores with mean 80 and std 10
scores_after = scores_before + np.random.normal(5, 5, 100)  # Adding random improvement (mean 5, std 5)

# Perform the paired t-test
t_stat, p_value = stats.ttest_rel(scores_before, scores_after)  # Paired t-test between before and after scores

# Print the results
print(f"T-statistic: {t_stat:.4f}")  # Print t-statistic
print(f"P-value: {p_value:.4f}")    # Print p-value

# Interpret the p-value
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("Reject the null hypothesis: The study program is effective.")
else:
    print("Fail to reject the null hypothesis: The study program is not effective.")

# Visualize the data
plt.hist(scores_before, bins=30, alpha=0.6, label='Before')  # Histogram for scores before
plt.hist(scores_after, bins=30, alpha=0.6, label='After')    # Histogram for scores after
plt.legend()  # Add legend to the plot
plt.xlabel('Score')  # Label for x-axis
plt.ylabel('Frequency')  # Label for y-axis
plt.title('Distribution of Scores Before and After the Program')  # Title of the plot
plt.show()  # Show the plot
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to simulate Type I and Type II errors
def simulate_hypothesis_tests():
    # Simulate a scenario that results in Type I error
    # Null hypothesis (H0): mean = 0 (true)
    # Alternative hypothesis (H1): mean != 0 (false)
    
    np.random.seed(0)
    
    # Scenario 1: Type I Error (Rejecting a true null hypothesis)
    true_mean = 0  # True null hypothesis mean
    sample_size = 30
    alpha_type1 = 0.05  # Significance level for the test
    
 # Generate data from the true null hypothesis distribution (mean = 0)
    data_type1 = np.random.normal(loc=true_mean, scale=1, size=sample_size)
    
    # Perform a two-tailed t-test (test if the sample mean is significantly different from 0)
    t_stat_type1, p_value_type1 = stats.ttest_1samp(data_type1, 0)
    
    print(f"Type I Error Scenario:")
    print(f"t-statistic: {t_stat_type1:.4f}, p-value: {p_value_type1:.4f}")
    
    # Reject null hypothesis if p-value < alpha
    if p_value_type1 < alpha_type1:
        print("Type I Error: Rejecting a true null hypothesis (False Positive).\n")
    else:
        print("No Type I Error: Failed to reject the true null hypothesis.\n")
    
     # Scenario 2: Type II Error (Failing to reject a false null hypothesis)
    # Null hypothesis (H0): mean = 0 (false)
    # Alternative hypothesis (H1): mean != 0 (true)
    
    # Scenario with Type II Error
    alternative_mean = 5  # A mean different from 0, H0 is false
    alpha_type2 = 0.05  # Significance level for the test
    
    # Generate data from the false null hypothesis distribution (mean = 5)
    data_type2 = np.random.normal(loc=alternative_mean, scale=1, size=sample_size)
    
    # Perform a two-tailed t-test (test if the sample mean is significantly different from 0)
    t_stat_type2, p_value_type2 = stats.ttest_1samp(data_type2, 0)
    
    print(f"Type II Error Scenario:")
    print(f"t-statistic: {t_stat_type2:.4f}, p-value: {p_value_type2:.4f}")
    
    # Fail to reject the null hypothesis if p-value > alpha (Type II error happens here)
    if p_value_type2 > alpha_type2:
        print("Type II Error: Failing to reject a false null hypothesis (False Negative).\n")
    else:
        print("No Type II Error: Rejected the false null hypothesis.\n")
    
    # Visualize the scenarios using histograms
    plt.figure(figsize=(12, 6))
    
    # Plot Type I error scenario
    plt.subplot(1, 2, 1)
    plt.hist(data_type1, bins=15, color='blue', alpha=0.6, label='Data (True Mean=0)')
    plt.axvline(np.mean(data_type1), color='red', linestyle='dashed', label=f"Sample Mean = {np.mean(data_type1):.2f}")
    plt.title("Scenario: Type I Error (True Mean = 0)")
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

     # Plot Type II error scenario
    plt.subplot(1, 2, 2)
    plt.hist(data_type2, bins=15, color='green', alpha=0.6, label='Data (True Mean=5)')
    plt.axvline(np.mean(data_type2), color='red', linestyle='dashed', label=f"Sample Mean = {np.mean(data_type2):.2f}")
    plt.title("Scenario: Type II Error (True Mean = 5)")
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Call the function to simulate the hypothesis tests
simulate_hypothesis_tests()
```

```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create a dataset of years of experience vs. salary
np.random.seed(0)  # For reproducibility
years_of_experience = np.random.uniform(0, 10, 100)
salary = 50000 + 10000 * years_of_experience + np.random.normal(0, 10000, 100)  # Base salary + noise

# Reshape the data for the linear regression model (required by sklearn)
years_of_experience = years_of_experience.reshape(-1, 1)

# Fit the linear regression model
model = LinearRegression()
model.fit(years_of_experience, salary)

# Print the coefficients of the model (intercept and slope)
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")

# Predict the salary for 5 years of experience
years_of_experience_new = np.array([[5]])
salary_new = model.predict(years_of_experience_new)
print(f"Predicted salary for 5 years of experience: ${salary_new[0]:.2f}")

# Calculate the R-squared value to assess the model's performance
r_squared = model.score(years_of_experience, salary)
print(f"R-squared: {r_squared:.4f}")

# Visualize the data and the regression line
plt.scatter(years_of_experience, salary, label='Data points', alpha=0.6)
plt.plot(years_of_experience, model.predict(years_of_experience), color='r', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```

```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Generate a dataset from a quadratic equation with noise
np.random.seed(0)  # For reproducibility
x = np.random.uniform(-10, 10, 100)
y = x**2 + np.random.normal(0, 10, 100)  # Quadratic relationship with noise

# Split the data into training and testing sets
train_size = int(0.8 * len(x))
x_train, x_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Fit a linear model (underfitting)
model_linear = LinearRegression()
model_linear.fit(x_train.reshape(-1, 1), y_train)

# Fit a high-degree polynomial (overfitting)
model_overfitting = make_pipeline(PolynomialFeatures(degree=10), LinearRegression())
model_overfitting.fit(x_train.reshape(-1, 1), y_train)

# Visualize both models and their performance on training and testing sets
x_range = np.linspace(-10, 10, 1000)  # Points for plotting the models
y_linear = model_linear.predict(x_range.reshape(-1, 1))  # Linear model prediction
y_overfitting = model_overfitting.predict(x_range.reshape(-1, 1))  # Overfitting model prediction

# Create the plot
plt.figure(figsize=(12, 6))

# Plot for training data
plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, label='Training data')
plt.plot(x_range, y_linear, label='Linear model', color='r')
plt.plot(x_range, y_overfitting, label='Overfitting model', color='g')
plt.legend()
plt.title('Training data')

# Plot for testing data
plt.subplot(1, 2, 2)
plt.scatter(x_test, y_test, label='Testing data')
plt.plot(x_range, y_linear, label='Linear model', color='r')
plt.plot(x_range, y_overfitting, label='Overfitting model', color='g')
plt.legend()
plt.title('Testing data')

# Show the plots
plt.show()
```

```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# Generate a synthetic dataset
np.random.seed(0)  # For reproducibility
x = np.random.uniform(0, 10, 100)  # Generate 100 data points uniformly distributed between 0 and 10
y = 2 * x + np.random.normal(0, 2, 100)  # Linear relation with some noise

# Split the data into training and testing sets (80% train, 20% test)
train_size = int(0.8 * len(x))
x_train, x_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape the data to meet the requirements of sklearn's models
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

# Linear regression model
model_linear = LinearRegression()
model_linear.fit(x_train, y_train)  # Fit the model on training data
y_pred_linear = model_linear.predict(x_test)  # Make predictions on test data
mse_linear = mean_squared_error(y_test, y_pred_linear)  # Calculate the MSE for linear regression

# Polynomial regression model (degree 2)
model_polynomial = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model_polynomial.fit(x_train, y_train)  # Fit the model on training data
y_pred_polynomial = model_polynomial.predict(x_test)  # Make predictions on test data
mse_polynomial = mean_squared_error(y_test, y_pred_polynomial)  # Calculate the MSE for polynomial regression

# Lasso regression model (L1 regularization)
model_lasso = Lasso(alpha=0.1)  # Lasso with a regularization strength of 0.1
model_lasso.fit(x_train, y_train)  # Fit the model on training data
y_pred_lasso = model_lasso.predict(x_test)  # Make predictions on test data
mse_lasso = mean_squared_error(y_test, y_pred_lasso)  # Calculate the MSE for lasso regression

# Print the mean squared errors of the models
print(f"Linear regression MSE: {mse_linear:.2f}")
print(f"Polynomial regression MSE: {mse_polynomial:.2f}")
print(f"Lasso regression MSE: {mse_lasso:.2f}")

# Visualize the data and the regression lines
plt.scatter(x, y, label='Data points')  # Plot the data points

# Plot the regression lines for each model
plt.plot(x_test, y_pred_linear, label='Linear regression', color='r')
plt.plot(x_test, y_pred_polynomial, label='Polynomial regression', color='g')
plt.plot(x_test, y_pred_lasso, label='Lasso regression', color='b')

# Adding labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Comparison of Regression Models')
plt.show()
```

```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate a synthetic dataset
np.random.seed(0)  # For reproducibility
x = np.random.uniform(0, 10, 100)  # Random x values between 0 and 10
y = 2 * x + np.random.normal(0, 2, 100)  # Linear relation with noise

# Reshape the data for the regression models
x = x.reshape(-1, 1)  # Reshape to column vector for sklearn

# Linear regression model
model_linear = LinearRegression()
model_linear.fit(x, y)  # Fit the linear model
y_pred_linear = model_linear.predict(x)  # Predict the y values using the linear model

# Polynomial regression model (degree 2)
model_polynomial = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model_polynomial.fit(x, y)  # Fit the polynomial model
y_pred_polynomial = model_polynomial.predict(x)  # Predict the y values using the polynomial model

# Visualize the data and the regression lines
plt.scatter(x, y, label='Data points')  # Scatter plot for the actual data
plt.plot(x, y_pred_linear, label='Linear regression', color='r')  # Red line for linear regression
plt.plot(x, y_pred_polynomial, label='Polynomial regression', color='g')  # Green curve for polynomial regression
plt.legend()  # Add a legend
plt.xlabel('X')  # X-axis label
plt.ylabel('Y')  # Y-axis label
plt.title('Linear vs Polynomial Regression')  # Title for the plot
plt.show()  # Display the plot
```

```python
# Import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Generate a synthetic dataset
np.random.seed(0)  # For reproducibility
hours_studied = np.random.uniform(0, 10, 100)
marks_scored = 2 * hours_studied + np.random.normal(0, 2, 100)

# Reshape the data for the regression model
hours_studied = hours_studied.reshape(-1, 1)

# Calculate the Pearson correlation coefficient
correlation_coefficient, _ = pearsonr(hours_studied.flatten(), marks_scored)
print(f"Pearson correlation coefficient: {correlation_coefficient:.2f}")

# Fit the linear regression model
model = LinearRegression()
model.fit(hours_studied, marks_scored)

# Predict the values
marks_scored_pred = model.predict(hours_studied)

# Visualize the data and the regression line
plt.scatter(hours_studied, marks_scored)
plt.plot(hours_studied, marks_scored_pred, color='r')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Linear Regression Model')
plt.show()
```
