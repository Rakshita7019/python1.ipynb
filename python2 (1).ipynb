```python
import random

def coin_toss_simulation(num_trials):
    heads_count = 0
    tails_count = 0

    for _ in range(num_trials):
        outcome = random.choice(["Heads", "Tails"])
        if outcome == "Heads":
            heads_count += 1
        else:
            tails_count += 1

    experimental_probability_heads = heads_count / num_trials
    experimental_probability_tails = tails_count / num_trials

    print(f"Experimental Probability of Heads: {experimental_probability_heads:.4f}")
    print(f"Experimental Probability of Tails: {experimental_probability_tails:.4f}")

def dice_roll_simulation(num_trials):
    sum_7_count = 0

    for _ in range(num_trials):
        dice1 = random.randint(1, 6)
        dice2 = random.randint(1, 6)
        if dice1 + dice2 == 7:
            sum_7_count += 1

    experimental_probability_sum_7 = sum_7_count / num_trials

    print(f"Experimental Probability of getting a sum of 7: {experimental_probability_sum_7:.4f}")

num_trials = 10000
print("Coin Toss Simulation:")
coin_toss_simulation(num_trials)
print("\nDice Roll Simulation:")
dice_roll_simulation(num_trials)
```

```python
import random

def estimate_probability(num_trials):
    successful_trials = 0

    for _ in range(num_trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:  
            successful_trials += 1

    estimated_probability = successful_trials / num_trials

    print(f"Estimated Probability of getting at least one '6' in 10 rolls: {estimated_probability:.4f}")

num_trials = 10000
estimate_probability(num_trials)
```

```python
import random

def draw_ball():
    balls = ['red'] * 5 + ['green'] * 7 + ['blue'] * 8
    return random.choice(balls)

def estimate_conditional_probability(num_trials):
    blue_then_red_count = 0
    blue_count = 0

    for _ in range(num_trials):
        ball1 = draw_ball()
        ball2 = draw_ball()
        if ball1 == 'blue':
            blue_count += 1
            if ball2 == 'red':
                blue_then_red_count += 1

    estimated_probability = blue_then_red_count / blue_count if blue_count != 0 else 0

    print(f"Estimated Probability of drawing a red ball given that the previous ball was blue: {estimated_probability:.4f}")

def verify_bayes_theorem(num_trials):
    red_count = 0
    blue_count = 0
    blue_then_red_count = 0

    for _ in range(num_trials):
        ball1 = draw_ball()
        ball2 = draw_ball()
        if ball1 == 'blue':
            blue_count += 1
            if ball2 == 'red':
                blue_then_red_count += 1
        if ball2 == 'red':
            red_count += 1

    estimated_probability_red_given_blue = blue_then_red_count / blue_count if blue_count != 0 else 0
    estimated_probability_red = red_count / num_trials
    estimated_probability_blue = blue_count / num_trials

    print(f"Estimated Probability of drawing a red ball: {estimated_probability_red:.4f}")
    print(f"Estimated Probability of drawing a blue ball: {estimated_probability_blue:.4f}")
    print(f"Estimated Probability of drawing a red ball given that the previous ball was blue (Bayes' theorem): {estimated_probability_red_given_blue:.4f}")

    bayes_theorem_result = (blue_then_red_count / num_trials) / (blue_count / num_trials) if blue_count != 0 else 0

    print(f"Bayes' theorem result: {bayes_theorem_result:.4f}")

num_trials = 100000
print("Estimating conditional probability:")
estimate_conditional_probability(num_trials)
print("\nVerifying Bayes' theorem:")
verify_bayes_theorem(num_trials)
```

```python
import random
import pandas as pd

def simulate_process(num_trials):
    data = {
        'A': [random.choice([0, 1]) for _ in range(num_trials)],
        'B': [random.choice([0, 1]) for _ in range(num_trials)]
    }
    df = pd.DataFrame(data)

    return df

def compute_conditional_probabilities(df):
    p_a_given_b = df[df['B'] == 1]['A'].mean() if df[df['B'] == 1].shape[0] > 0 else 0
    p_b_given_a = df[df['A'] == 1]['B'].mean() if df[df['A'] == 1].shape[0] > 0 else 0

    return p_a_given_b, p_b_given_a

num_trials = 100000
df = simulate_process(num_trials)
p_a_given_b, p_b_given_a = compute_conditional_probabilities(df)

print(f"P(A|B): {p_a_given_b:.4f}")
print(f"P(B|A): {p_b_given_a:.4f}")
```

```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

np.random.seed(0)  
sample = np.random.choice(values, size=1000, p=probabilities)

empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)
empirical_std_dev = np.std(sample)

print(f"Empirical Mean: {empirical_mean:.4f}")
print(f"Empirical Variance: {empirical_variance:.4f}")
print(f"Empirical Standard Deviation: {empirical_std_dev:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

np.random.seed(0)  
samples = np.random.exponential(scale=5, size=2000)

plt.hist(samples, bins=30, density=True, alpha=0.6, label='Histogram', color='blue')

x = np.linspace(0, 20, 100)  
y = expon.pdf(x, scale=5)  
plt.plot(x, y, 'r-', label='PDF', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution with Mean 5')

plt.legend()

plt.show()
```
